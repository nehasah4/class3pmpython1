```python
from bs4 import BeautifulSoup
import requests
url  = BeautifulSoup('https://www.worldometers.info/coronavirus/','html.parser')
soup = requests.get(url)
soup


```

```python
codes = BeautifulSoup(soup.text,"lxml")
codes
```

```python
codes.table
```

```python
table_code = codes.table
table_code
```

```python
tags = table_code.find_all('tr')
tags
```

```python
for tag in tags:
    print(tag.text)
```

```python
list_data =[]
for tag in tags:
    
    list_data.append(tag.text)
    
    
list_data
```

```python
list_data =[]
for tag in tags:
    list_data.append(tag.text.split('\n'))
    
    
list_data
```

```python
import csv
file = open('covid.csv','w')
x = csv.writer(file)
x.writerows(list_data)
```

```python
tags = table_code.find_all('tr')
```
